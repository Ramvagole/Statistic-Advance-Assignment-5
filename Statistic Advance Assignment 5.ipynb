```python
#Q1):-
To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python, you can utilize the scipy.stats module. Here's how you can do it:
python

import scipy.stats as stats

sample_mean = 50
sample_std = 5
sample_size = # Specify the sample size here
confidence_level = 0.95
standard_error = sample_std / (sample_size ** 0.5)
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, sample_size - 1) * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error
print("95% Confidence Interval:")
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)

Make sure to replace sample_size with the actual size of your sample.

Interpreting the results: The 95% confidence interval is a range of values that we can be 95% confident to contain the true population mean. In this case, if we were to take multiple samples from the population and calculate their 95% confidence intervals using the same method, approximately 95% of those intervals would contain the true population mean.

For example, if the calculated confidence interval is (47.5, 52.5), it means we are 95% confident that the true population mean falls within this range. In other words, we can say with 95% confidence that the population mean is between 47.5 and 52.5 based on the given sample data.
```

```python
#Q2):-
import numpy as np
import scipy.stats as stats
observed_freq = np.array([50, 45, 55, 30, 25, 45])
expected_freq = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_freq)
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)
alpha = 0.05
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("The distribution of colors in the bag does not match the expected distribution.")
else:
    print("The distribution of colors in the bag matches the expected distribution.")
```

```python
#Q4):-
import scipy.stats as stats

total_population = 500
smokers = 60

confidence_level = 0.95
sample_proportion = smokers / total_population
standard_error = (sample_proportion * (1 - sample_proportion) / total_population) ** 0.5
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * standard_error
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error
print("95% Confidence Interval:")
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
```

```python
#Q6):-
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
df = 10
x = np.linspace(0, 30, 1000)
pdf = stats.chi2.pdf(x, df)
plt.plot(x, pdf, label=f"df = {df}")
x_shade = np.linspace(15, 30, 1000)
pdf_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, pdf_shade, color='gray', alpha=0.5)
plt.xlabel("Chi-square Statistic")
plt.ylabel("Probability Density Function (PDF)")
plt.legend()
plt.show()
```

```python
#Q7):-
import statsmodels.api as sm

total_sample = 1000
coke_preferred = 520
sample_proportion = coke_preferred / total_sample
standard_error = (sample_proportion * (1 - sample_proportion) / total_sample) ** 0.5
confidence_level = 0.99
margin_of_error = sm.stats.proportion.proportion_confint(coke_preferred, total_sample, alpha=(1-confidence_level))
lower_bound = sample_proportion - margin_of_error[0]
upper_bound = sample_proportion + margin_of_error[1]
print("99% Confidence Interval:")
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
```

```python
#Q8):-
import numpy as np
import scipy.stats as stats
observed_freq = np.array([45, 55])
expected_freq = np.array([50, 50])
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)
alpha = 0.05
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("The observed frequencies significantly differ from the expected frequencies.")
else:
    print("The observed frequencies do not significantly differ from the expected frequencies.")
```

```python
#Q9):-
import numpy as np
import scipy.stats as stats
observed = np.array([[60, 140], [30, 170]])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)
alpha = 0.05
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
if p_value < alpha:
    print("There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("There is no significant association between smoking status and lung cancer diagnosis.")
```

```python
#Q11):-
import numpy as np
import scipy.stats as stats
sample_mean = 72
sample_std = 10
sample_size = 30
null_mean = 70
t_stat, p_value = stats.ttest_1samp(np.random.normal(sample_mean, sample_std, sample_size), null_mean)
alpha = 0.05
print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < alpha:
    print("The population mean is significantly different from 70.")
else:
    print("The population mean is not significantly different from 70.")
```
